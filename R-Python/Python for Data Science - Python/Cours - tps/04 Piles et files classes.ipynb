```python
# Les structures de données complexes
# Implémentation d'une Pile, File à l'aide d'une Liste
 
class Pile:
    """Implémentation d'une Pile à l'aide d'une Liste"""
    def __init__(self):
        # Initialiser une pile vide (libre)
        self.pile = []
    def SOM_P(self):
        """Retourner le sommet de la pile"""
        try:
            return self.pile[-1]
        except Exception as e:
            print("Error:", e)
    def Empiler(self, element):
        """Insérer un élément au sommet de la pile (Push)"""
        return self.pile.append(element)
    def Depiler(self):
        """Supprimer l’élément de sommet de la pile (Pop)"""
        try:
            return self.pile.pop()
        except Exception as e:
            print("Error:", e)
    def PileVide(self):
        """Retourner Vrai si la pile est vide, sinon Faux"""
        if self.pile == []:
            return True
        else:
            return False
 
    def AfficherPile(self):
        """Afficher tous les éléments de la Pile"""
        for i in self.pile:
            print(self.pile[i])
 
 
class File:
     
    """Implémentation d'une File à l'aide d'une Liste"""
 
    def __init__(self):
        # Initialiser une file vide (libre)
        self.file = []
 
    def Enfiler(self, element):
        """Insérer un élément à la queue d’une file"""
        return self.file.append(element)
 
    def Defiler(self):
        """Supprimer l’élément de la tete une file"""
        try:
            return self.file.pop(0)
        except Exception as e:
            print("Error:", e)
 
    def AfficherFile(self):
        """Afficher tous les éléments de la file"""
        for i in self.file:
            print(self.file[i])
        
```
